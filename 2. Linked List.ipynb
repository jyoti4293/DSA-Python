Linked List: Linear data structure having collection of nodes stored at non-continuous memory location.

Empty LL: Head= None,
Len(LL): Number of nodes

Operations on LL:
1. Insert- head, tail, middle
2. Traverse- print
3. Delete- head, tail, value, index
4. Search- value, index


Array vs Linked List:

Read (Traverse/Indexing) operations are faster with array, write (Insert/Delete) operations are faster with linked list

```python
class Node:
    
    def __init__(self,value):
        self.value=value
        self.next=None
```
