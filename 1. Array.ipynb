Array- Linear data structure that contains objects of single data type stored in continuous memory location.

Disadvantages- 
1. Fixed size 
2. Can only store single data type

Referential Array- Call by reference- stores the memory location of the objects and object can be called by reference instead of call by value.

Dynamic Array- 
1. Size of the array is not fixed and can be increased when required. 
2. Can be heterogenous- can contains object of mixed data types.

Python list is a dynamic array. To understand the concept of dynamic array- take a c type (static) array and
make it dynamic along with all the properies of python list.

```python
#create class list 

class MyList():
    # My list can be identified by size and elements- size is the max size of the list at that time
    # n is the number is elements present in the list at that time
    # A is the c-static array
    
    def __init__(self):
        self.size=1
        self.n=0
        self.A= self.make_array(self.size)
        
    def make_array(self,capacity):
        # creates a c type fixed size referential array
        return (capacity*ctypes.py_object)()
    
    def __len__(self):       
        return self.n
    
    def __str__(self):
        result=''
        for i in range (self.n):
            result = result + str(self.A[i]) +','
            
        return '['+ result[:-1]+ ']'
    
    def __getitem__(self, index):
        if 0<=index<self.n:
            return self.A[index]
        else:
            return "Index Error"
    
    def append(self,element):
        if self.size==self.n:
            #resize
            self.resize(self.size*2)
            pass
        #append
        self.A[self.n]=element
        self.n=self.n+1
    
    def resize(self, new_capacity):
        # create new array with new capacity
        B=self.make_array(new_capacity)
        self.size= new_capacity
        
        #copy the content of A to B
        for i in range(self.n):
            B[i]=self.A[i]
        
        # reassign A
        self.A=B 
        
    def pop(self):
        if self.n==0:
            return "Empty List"
        
        print(self.A[self.n-1])
        self.n= self.n-1
    
    def clear(self):
        self.n=0
        self.size=1
        
    def find(self, item):
        
        for i in range(self.n):
            if self.A[i]==item:
                return i
        return "Value Error"
    
    def insert(self, pos, item):
        
        if self.size==self.n:
            self.resize(self.size*2)
            
        for i in range(self.n, pos,-1):
            self.A[i]=self.A[i-1]
        
        self.A[pos]=item
        self.n= self.n+1
    
    def remove(self, item):
        
        index=self.find(item)
        
        if isinstance(index, int):
            for i in range(index, self.n-1):
                self.A[i]=self.A[i+1]
            
            self.n=self.n-1
        else:
            return "Value Error"
    
    def sum(self, list):
        
        #resize
        self.resize(self.size+len(list))
        
        for i in range(len(list)):
            self.append(list[i])
        
```
